# Bollinger Fibonacci Strategy Demo

This notebook demonstrates the **strategy-oriented** Bollinger Fibonacci endpoint used by the app:
- `POST /bollinger_fibonacci`
- MCP tool: `analyze_bollinger_fibonacci_performance`

It requests a strategy report and prints key metrics/trades if available in the structured payload.

```python
import json
import requests
```

```python
API_URL = "http://localhost:8000/bollinger_fibonacci"
payload = {
    "symbol": "AAPL",
    "period": "1y",
    "window": 20,
    "num_std": 2,
    "window_swing_points": 10,
    "agent_type": "adk_agentic"
}
payload
```

```python
response = requests.post(API_URL, json=payload, timeout=300)
print("Status:", response.status_code)
response.raise_for_status()
result = response.json()
list(result.keys())
```

```python
print("Analysis Type:", result.get("analysis_type"))
print("Symbol:", result.get("symbol"))
print("Duration (s):", result.get("duration_seconds"))
print("Tools Approach:", result.get("tools_approach"))
```

```python
timeseries = result.get("timeseries") or []
print("Trades Returned:", len(timeseries))
timeseries[:5]
```

```python
report_text = result.get("report", "")
print(report_text[:3000])
```
