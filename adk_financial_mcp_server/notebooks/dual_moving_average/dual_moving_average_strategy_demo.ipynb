# Dual Moving Average Strategy Demo

This notebook calls the strategy-oriented endpoint used by the app:
- `POST /dual_moving_average`
- MCP tool: `analyze_dual_ma_strategy`

It validates response metadata and prints the report preview.

```python
import requests
```

```python
API_URL = "http://localhost:8000/dual_moving_average"
payload = {
    "symbol": "AAPL",
    "period": "5y",
    "short_period": 50,
    "long_period": 200,
    "ma_type": "SMA",
    "agent_type": "adk_agentic"
}
payload
```

```python
response = requests.post(API_URL, json=payload, timeout=300)
print("Status:", response.status_code)
response.raise_for_status()
result = response.json()
list(result.keys())
```

```python
print("Analysis Type:", result.get("analysis_type"))
print("Symbol:", result.get("symbol"))
print("Duration (s):", result.get("duration_seconds"))
print("Tools Approach:", result.get("tools_approach"))
```

```python
timeseries = result.get("timeseries") or []
print("Timeseries trades returned:", len(timeseries))
timeseries[:5]
```

```python
report_text = result.get("report", "")
print(report_text[:3000])
```
